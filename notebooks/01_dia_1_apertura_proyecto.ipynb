```python
from pathlib import Path

project_root = Path.cwd().resolve()

if project_root.name == "notebooks":
    project_root = project_root.parent

folders = [
    project_root / "data" / "raw",
    project_root / "data" / "processed",
    project_root / "notebooks"
]

for folder in folders:
    folder.mkdir(parents=True, exist_ok=True)

print("Estructura base verificada.")
print(f"Carpeta raíz del proyecto: {project_root}")

for folder in folders:
    print(f"- {folder}")
```

```python
from pathlib import Path

data_file = project_root / "data" / "raw" / "retail_sales_dataset.xlsx"

if data_file.exists():
    print("Dataset encontrado correctamente.")
    print(data_file)
else:
    print("No se encontró el dataset en la ruta esperada.")
    print("Coloca el archivo 'retail_sales_dataset.xlsx' dentro de 'data/raw/'.")
```

```python
import pandas as pd
from IPython.display import display

expected_sheets = {"Customers", "Products", "Stores", "Transactions"}

if data_file.exists():
    xls = pd.ExcelFile(data_file)
    detected_sheets = set(xls.sheet_names)

    print("Hojas detectadas:")
    print(xls.sheet_names)

    missing_sheets = expected_sheets - detected_sheets
    extra_sheets = detected_sheets - expected_sheets

    if not missing_sheets:
        print("\nTodas las hojas esperadas están presentes.")
    else:
        print("\nFaltan estas hojas:")
        print(sorted(missing_sheets))

    if extra_sheets:
        print("\nHojas adicionales detectadas:")
        print(sorted(extra_sheets))

    summary_rows = []

    for sheet_name in xls.sheet_names:
        preview = pd.read_excel(data_file, sheet_name=sheet_name, nrows=5)
        summary_rows.append(
            {
                "hoja": sheet_name,
                "num_columnas": preview.shape[1],
                "columnas": ", ".join(preview.columns)
            }
        )

    display(pd.DataFrame(summary_rows))
else:
    print("Primero coloca el dataset en 'data/raw/' para continuar.")
```
